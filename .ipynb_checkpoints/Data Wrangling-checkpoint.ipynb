````python
#importing required modules
import pandas as pd
import numpy as np
````



### 1. Examine the data frame for duplicates (based on ID), and drop any duplicates that exist. 


````python
p_df = pd.read_csv('data/personality_scores.csv', sep=';')
p_df.head()
````

````python
# finding the null values
p_df.isnull().sum()
````

````python
#checking for duplicates

p_df['ID'].is_unique
````

### There are no duplicates, if we had found any we could use the function `duplicate_removal` below to remove them.

````python
def duplicate_removal():
    first = len(p_df.ID)
    duplicates = len(p_df.drop_duplicates(subset='ID', keep='first'))
    return first - duplicates
duplicate_removal()
````

### Use an assert statement to check that the new data frame is the length of the unique entries of the original data frame.

Since there are no duplicates the original data has not been changed. We could check this with;
```

```

````python
# setting index to be "ID" column in the data frame

p_df = p_df.set_index('ID')

p_df.head()
# p_df.shape
````

````python
# dropping of null values as part of the cleaning process

p_df = p_df.dropna(axis = 1)
p_df.head()
````

### Create new columns containing the total score of each of the personality test subscales.

````python
# subscaling the required columns and getting the totals for the new columns
def total_score(score, subscale):
    x = sum([eval(score[i])[1] for i in range(50) if eval(score[i])[0] == subscale])
    return(x)
# x is the total of each of the subscores
````

### Appending the newly created columns to the data frame 

````python
# adding the new columns 
p_df['Extraversion'] = p_df.apply(lambda row: total_score(row, 1), axis=1)
p_df['Agreeableness'] = p_df.apply(lambda row: total_score(row, 2), axis=1)
p_df['Conscientiousness'] = p_df.apply(lambda row: total_score(row, 3), axis=1)
p_df['Emotional Stability/Neuroticism'] = p_df.apply(lambda row: total_score(row, 4), axis=1)
p_df['Intellect/Imagination'] = p_df.apply(lambda row: total_score(row, 5), axis=1)
````

````python
p_df.head()
````

### 4.Filter the merged data frame so that you get only the applicants who scored less than 30 on emotional stability, conscientiousness AND agreeableness. Print the ID numbers and departments of these applicants to the screen, and also assign these applicants the tag “high_risk” in a new column. All other applicants get the tag “low_risk”

````python
full_df.head()
````

````python
# combined score of less than 30 from emotional stability, conscientiousness AND agreeableness. 
# define the criteria
criteria = 30
# get the variable that is less than the given criteria in all three mentioned above 
full_df[(full_df['Emotional Stability/Neuroticism'] < criteria) & \
        (full_df.Conscientiousness < criteria) & (full_df.Agreeableness < criteria)]
````

````python
# define what high_risk and low_risk is

full_df['Risk'] = np.where(
    (full_df['Emotional Stability/Neuroticism'] < criteria) & \
    (full_df.Conscientiousness < criteria) & (full_df.Agreeableness < criteria), 'High risk', 'Low risk'
)
full_df.head()
````

### A table showing the sum of Risk according to their departments 

````python
dff = full_df[['Department', 'Risk']].groupby(['Department', 'Risk']).size()
pd.DataFrame(dff)
````
