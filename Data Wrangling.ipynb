### Checking and removal of duplicates

````python
# checking for duplicates
p_df['ID'].is_unique

#removal of duplicates
def duplicate_removal():
    first = len(p_df.ID)
    duplicates = len(p_df.drop_duplicates(subset='ID', keep='first'))
    return first - duplicates
duplicate_removal()
````

````python
p_df = p_df.set_index('ID')

p_df.head()
# p_df.shape
````

````python
p_df = p_df.dropna(axis = 1)
p_df.head()
````

### Create new columns containing the total score of each of the personality test subscales.

````python
# calculating the total score
def total_score(score, subscale):
    x = sum([eval(score[i])[1] for i in range(50) if eval(score[i])[0] == subscale])
    return(x)
# total_score()
````

````python
# Creating of new columns
p_df['Extraversion'] = p_df.apply(lambda row: total_score(row, 1), axis=1)
p_df['Agreeableness'] = p_df.apply(lambda row: total_score(row, 2), axis=1)
p_df['Conscientiousness'] = p_df.apply(lambda row: total_score(row, 3), axis=1)
p_df['Neuroticism'] = p_df.apply(lambda row: total_score(row, 4), axis=1)
p_df['Intellect'] = p_df.apply(lambda row: total_score(row, 5), axis=1)
````

### Merging the data frames
